```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix

data = pd.read_csv("/content/Fish-Market/Fish.csv")

bream_data = data[data['Species'] == 'Bream']
smelt_data = data[data['Species'] == 'Smelt']

bream_length = bream_data['Length2'].values.tolist()
bream_weight = bream_data['Weight'].values.tolist()
smelt_length = smelt_data['Length2'].values.tolist()
smelt_weight = smelt_data['Weight'].values.tolist()

print(bream_length)
print(bream_weight)
print(smelt_length)
print(smelt_weight)
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length,bream_weight)
plt.xlabel('length')
plt.ylabel('weigth')
plt.show()
```

```python
plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length=[]
weight=[]
length= bream_length + smelt_length
weight=bream_weight + smelt_weight

print(length)
print(weight)
```

```python
fish_data=[[l,w] for l,w in zip(length,weight)]
print(fish_data)

```

```python
fish_target=[1]*35+[0]*14 #print(len(fish_data))하니 개수가 달라서 보정 후 진행
print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
```

```python
kn.fit(fish_data,fish_target)
kn.score(fish_data,fish_target)
```

```python
plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)

plt.scatter(15,80,marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 확인문제 4.
kn=KNeighborsClassifier()
kn.fit(fish_data,fish_target)

for n in range(5,50):
  kn.n_neighbors=n
  score=kn.score(fish_data,fish_target)
  if score<1:
    print(n,score)
    break
```

```python
roach_data = data[data['Species'] == 'Roach']
whitefish_data = data[data['Species'] == 'Whitefish']
parkki_data = data[data['Species'] == 'Parkki']
perch_data = data[data['Species'] == 'Perch']
pike_data = data[data['Species'] == 'Pike']

roach_length=roach_data['Length2'].values.tolist()
roach_weight=roach_data['Weight'].values.tolist()
whitefish_length=whitefish_data['Length2'].values.tolist()
whitefish_weight=whitefish_data['Weight'].values.tolist()
parkki_length=parkki_data['Length2'].values.tolist()
parkki_weight=parkki_data['Weight'].values.tolist()
perch_length=perch_data['Length2'].values.tolist()
perch_weight=perch_data['Weight'].values.tolist()
pike_length=pike_data['Length2'].values.tolist()
pike_weight=pike_data['Weight'].values.tolist()

data=[roach_data,whitefish_data,parkki_data,perch_data,pike_data]
length=[roach_length,whitefish_length,parkki_length,perch_length,pike_length]
weight=[roach_weight,whitefish_weight,parkki_weight,perch_weight,pike_weight]

for l,w,d in zip(length,weight,data):
  ll=bream_length + l
  ww=bream_weight + w

  fish_data=[[l,w] for l,w in zip(ll,ww)]
  fish_target=[1]*35+[0]*(len(d))

  kn=KNeighborsClassifier()
  kn.fit(fish_data,fish_target)

  plt.scatter(bream_length,bream_weight)
  plt.scatter(l,w)
  plt.scatter(30,600,marker='^')
  plt.xlabel('length')
  plt.ylabel('weight')
  plt.show()
```
