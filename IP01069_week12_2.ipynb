# **가중치 시각화**

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
(train_input,train_target),(test_input,test_target)=keras.datasets.fashion_mnist.load_data()
train_scaled=train_input.reshape(-1,28,28,1)/255.0
train_scaled,val_scaled,train_target,val_target=train_test_split(train_scaled,train_target,test_size=0.2,random_state=42)
```

```python
model=keras.Sequential()
model.add(keras.layers.Conv2D(32,kernel_size=3,activation='relu',padding='same',input_shape=(28,28,1)))
model.add(keras.layers.MaxPooling2D(2))
model.add(keras.layers.Conv2D(64,kernel_size=(3,3),activation='relu',padding='same'))
model.add(keras.layers.MaxPooling2D(2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dropout(0.4))
model.add(keras.layers.Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

checkpoint_cb=keras.callbacks.ModelCheckpoint('best-cnn-model.h5',save_best_only=True)
early_stopping_cb=keras.callbacks.EarlyStopping(patience=2,restore_best_weights=True)
history=model.fit(train_scaled,train_target,epochs=20,validation_data=(val_scaled,val_target),callbacks=[checkpoint_cb,early_stopping_cb])
```

# **응용 예제**
앞에서 구성한 숫자 MNIST에 대해서 Conv2D 모델에 대해 가중치시각화와 특성 시각화를 수행하시오.

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

(mnist_train_input, mnist_train_target), (mnist_test_input, mnist_test_target) = keras.datasets.mnist.load_data()

mnist_train_scaled = mnist_train_input.reshape(-1, 28, 28, 1) / 255.0
mnist_test_scaled = mnist_test_input.reshape(-1, 28, 28, 1) / 255.0

mnist_train_scaled, mnist_val_scaled, mnist_train_target, mnist_val_target = train_test_split(
    mnist_train_scaled, mnist_train_target, test_size=0.2, random_state=42
)

print(f"MNIST training data shape: {mnist_train_scaled.shape}")
print(f"MNIST validation data shape: {mnist_val_scaled.shape}")
print(f"MNIST test data shape: {mnist_test_scaled.shape}")
```

```python
mnist_model = keras.Sequential()
mnist_model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1)))
mnist_model.add(keras.layers.MaxPooling2D(2))
mnist_model.add(keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'))
mnist_model.add(keras.layers.MaxPooling2D(2))
mnist_model.add(keras.layers.Flatten())
mnist_model.add(keras.layers.Dense(100, activation='relu'))
mnist_model.add(keras.layers.Dropout(0.4))
mnist_model.add(keras.layers.Dense(10, activation='softmax'))

mnist_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

checkpoint_cb_mnist = keras.callbacks.ModelCheckpoint('best-cnn-mnist-model.h5', save_best_only=True)
early_stopping_cb_mnist = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

history_mnist = mnist_model.fit(
    mnist_train_scaled, mnist_train_target,
    epochs=20,
    validation_data=(mnist_val_scaled, mnist_val_target),
    callbacks=[checkpoint_cb_mnist, early_stopping_cb_mnist]
)
```

```python
loaded_mnist_model = keras.models.load_model('best-cnn-mnist-model.h5')

mnist_conv_layer = loaded_mnist_model.layers[0]

mnist_conv_weights = mnist_conv_layer.weights[0].numpy()

plt.hist(mnist_conv_weights.reshape(-1, 1), bins=50)
plt.xlabel('Weight Value')
plt.ylabel('Count')
plt.show()

fig, axs = plt.subplots(4, 8, figsize=(15, 7))

for i in range(4):
    for j in range(8):
        filter_idx = i * 8 + j
        if filter_idx < mnist_conv_weights.shape[-1]:
            axs[i, j].imshow(mnist_conv_weights[:, :, 0, filter_idx], vmin=-0.5, vmax=0.5, cmap='gray')
            axs[i, j].axis('off')
plt.show()
```

```python
loaded_mnist_model = keras.models.load_model('best-cnn-mnist-model.h5')
loaded_mnist_model.build(input_shape=(None, 28, 28, 1))

input_image_mnist = mnist_test_scaled[0:1]

input_tensor_mnist = keras.Input(shape=(28, 28, 1))

output_from_layer0_mnist = loaded_mnist_model.layers[0](input_tensor_mnist)
output_from_layer1_mnist = loaded_mnist_model.layers[1](output_from_layer0_mnist)
output_of_conv2_layer_mnist = loaded_mnist_model.layers[2](output_from_layer1_mnist)

mnist_feature_map_model = keras.Model(inputs=input_tensor_mnist, outputs=output_of_conv2_layer_mnist)

mnist_feature_maps = mnist_feature_map_model.predict(input_image_mnist)

plt.imshow(input_image_mnist[0, :, :, 0], cmap='gray_r')
plt.axis('off')
plt.show()

fig, axs = plt.subplots(8, 8, figsize=(12, 12))
for i in range(8):
    for j in range(8):
        axs[i, j].imshow(mnist_feature_maps[0, :, :, i * 8 + j], cmap='viridis')
        axs[i, j].axis('off')
plt.show()
```
